# Handling Outliers

```python
import numpy as np
```

```python
lst_marks = [45,67,89,34,23,56,78,89,99,90,23,56,78,59,60,67,90,54,21,66,77,88,99,1000,1100]
```

```python
np.percentile(lst_marks,[25])
```

```python
np.percentile(lst_marks,[75])
```

```python
minimum, Q1,Q2,Q3, maximum = np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum
```

```python
maximum
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)
```

```python
lower_fence
```

```python
higher_fence
```

```python
outliers = []
for i in lst_marks:
    if i>=6.5 and i<=138.5:
        print(i, end=',')
    else:
        outliers.append(i)
```
