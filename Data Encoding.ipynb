# Data Encoding

# 1. One Hot Encoding (OHE)

```python
import pandas as pd
```

```python
df = pd.DataFrame({'color':['red','blue','green','green','red','blue']})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['color']])
```

```python
encoder.fit_transform(df[['color']]).toarray() # albhabetically seen the binary vector
```

```python
encoded = encoder.fit_transform(df[['color']])
```

```python
encoded_df = pd.DataFrame(encoded.toarray(), columns = encoder.get_feature_names_out())
```

```python
pd.concat([df, encoded_df], axis = 1)
```

# 2. Label Encoding

```python
df = pd.DataFrame({'color':['red','blue','green','green','red','blue']})
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
encoded = encoder.fit_transform(df['color'])
```

```python
encoded_df = pd.DataFrame(encoded)
```

```python
encoded_df
```

```python
pd.concat([encoded_df, df], axis=1)
```

# 3. Ordinal Encoding

```python
df = pd.DataFrame({'size':['small','medium','large','medium','small','large']})
```

```python
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder(categories = [['small','medium','large']])
```

```python
encoded = encoder.fit_transform(df[['size']])
```

```python
encoded_df = pd.DataFrame(encoded)
```

```python
pd.concat([encoded_df, df], axis=1)
```

# 3. Target Guided Ordinal Encoding

```python
df = pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','Paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
df
```

```python
# calculate the mean price for each city
df.groupby('city')['price'].mean()
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
# replace each city with its mean price
df['city_encoded'] = df['city'].map(mean_price)
```
